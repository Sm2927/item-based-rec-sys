```python
train = np.copy(nprat)
test=np.zeros((ntu,ntm))
#z for iterate over each row

user=0
#print(nprat)
for row in nprat:
    p=np.nonzero(row) #placing the index of non-zero values in p
    
    split=int(len(p[0])*0.2) #we have to split non-zero index for 20% in every row
    rand =random.choice(p[0],split,replace=False) #generating random indices from p of size split(i.e, 20% of no. of non-zero values ineach row)
    for i in range(split):   #making the train and test sets
        test[user,rand[i]] =nprat[user,rand[i]]
        train[user,rand[i]] = 0 
    user += 1

```

```python
#adjusted cosine similarity calculation
def adjustedCos(train):
  
    userm = train.sum(axis=1)/(train !=0).sum(axis=1) #mean value of user ratings
    ratm = np.where((train !=0),train-userm[:,None],train) #subtracting the mean value from every non-zero rating
    item_sim=np.zeros((ntm,ntm)) #similarity matrix
    for i in range(ntm): #for every movie
        print(item_sim)
        print("\n")
        for j in range(i,ntm): 
            n=0
            d1=0
            d2=0
            ci=np.where((train[:,i] !=0) * (train[:,j]) )[0] #users who have rated both items i and j
            for k in ci:
                n=n+ratm[k][i] * ratm[k][j]
                d1=d1 + ratm[k][i]**2
                d2=d2 + ratm[k][j]**2
                item_sim[i,j] = n/sqrt(d1*d2 + 10** -12) #RuntimeWarning: invalid value encountered in double_scalars(division by 0)
                if(i!=j):
                    item_sim[j,i] = item_sim[i,j]  #as it is a similarity matrix
       
    
    return item_sim
    
```

```python
item_sim=adjustedCos(train)
```

```python
item_sim=np.where((item_sim <0),0,item_sim)   #changing negative values to 0

#finding prediction using weighted sum
a=np.dot(train,item_sim) #multiplying rating value with similarity value (train.shape =(6040, 3952)) (item_sim.shape=(3952,3952))
b=np.zeros((ntu,ntm))  
for i in range(ntu) : #for every user
    c=np.nonzero(train[i])  #taking index of non-zero ratings
    for j in range(ntm):
        b[i,j] = item_sim[j,c].sum() 
predm=a/b
predm
```
